```python
def format_tract_to_6_digits(x):
    x = str(x)  # Convert to string to handle both integer and decimal cases
    if '.' in x:  # Check if there's a decimal point
        base, decimal = x.split('.')  # Split into base and decimal parts
        # Ensure the base is 4 digits and the decimal is 2 digits
        return f"{int(base):04d}{decimal[:2].ljust(2, '0')}"  # Ensure 6 digits
    else:
        # For whole number tracts (e.g., 7506), pad to 6 digits
        return f"{int(x):06d}"  # Pad to 6 digits

# Apply this function to your tract column
centroids['formatted_tract'] = centroids['tract'].apply(format_tract_to_6_digits)
```

```python
centroids
```

```python
# Remove the first two digits from the county code and pad to 3 digits if necessary
centroids['formatted_county'] = centroids['county'].astype(str).str[2:].str.zfill(3)
```

```python
centroids['GEOID'] = (
    centroids['state'].astype(str).str.zfill(2) +
    centroids['formatted_county'] +
    centroids['formatted_tract']
)
```

```python
tracts = merged[['GEOID', 'IntPtLat', 'IntPtLon', 'Category_Number']].copy()
```

```python
natural_gas = merged[merged['Category_Number'] == 2][['latitude', 'longitude']]
```

```python
from scipy.spatial import cKDTree
import numpy as np
```

```python
from scipy.spatial import cKDTree

# Step 1: Initialize Band column
tracts['Band'] = 0

# Step 2: Assign Band 1 to host tracts
tracts.loc[tracts['Category_Number'] == 2, 'Band'] = 1

# Step 3: Build KDTree from plant locations
plant_coords = natural_gas[['latitude', 'longitude']].to_numpy()
tree = cKDTree(plant_coords)

# Step 4: Filter non-host tracts with valid coordinates
mask_non_host = tracts['Band'] == 0
valid_coords = tracts['IntPtLat'].notna() & tracts['IntPtLon'].notna()
tract_coords = tracts.loc[mask_non_host & valid_coords, ['IntPtLat', 'IntPtLon']].to_numpy()

# Step 5: Compute distance to nearest plant
distances, _ = tree.query(tract_coords, k=1)
distances_km = distances * 111

# Step 6: Store distances temporarily
tracts.loc[mask_non_host & valid_coords, 'distance_to_plant'] = distances_km

# Step 7: Assign bands based on distance
tracts.loc[mask_non_host & (tracts['distance_to_plant'] <= 1), 'Band'] = 1  # Near-host tracts
tracts.loc[mask_non_host & (tracts['distance_to_plant'] > 1) & (tracts['distance_to_plant'] <= 5), 'Band'] = 2
tracts.loc[mask_non_host & (tracts['distance_to_plant'] > 5) & (tracts['distance_to_plant'] <= 10), 'Band'] = 3
tracts.loc[mask_non_host & (tracts['distance_to_plant'] > 10) & (tracts['distance_to_plant'] <= 20), 'Band'] = 4

# Optional: clean up
# tracts.drop(columns='distance_to_plant', inplace=True)
```

```python
# Drop any old Band or distance columns to avoid duplication
merged = merged.drop(columns=['Band', 'distance_to_plant'], errors='ignore')
```

```python
# Merge Band and distance info from tracts based on GEOID
bands = merged.merge(
    tracts[['GEOID', 'Band', 'distance_to_plant']],
    on='GEOID',
    how='left'
)
```

Band 1 is for tracts that host power plants or those whose distance from the tract's centroid to the closest power plant facility is greater than up to 1km.
Band 2 will be for tracts that do not host power plants but whose distance from the tract's centroid to the closest power plant facility is greater than 1 km and up to 5km.
Band 3 is for tracts that do not host power plants but whose distance from the tract's centroid to the closest power plant facility is greater than 5 km and up to 10 km.
Band 4 is for tracts that do not host power plants but whose distance from the tract's centroid to the closest power plant facility is greater than 10 km and up to 20 km.
