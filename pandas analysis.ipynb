```python
import pandas as pd
```

```python
df_plants = pd.read_csv('plants.csv')
```

```python
df_water = pd.read_csv('water_usage.csv')
```

```python
df_plants.head()
```

```python
df_plants.shape
```

```python
df_water.head()
```

```python
df_plants.rename(columns={'Plant Name': 'plant_name'}, inplace=True)
```

```python
df = pd.merge(df_plants, df_water, on='plant_name', how='left')
```

```python
df['Natural Gas'] = df['Primary Technology'].apply(lambda x: 'Natural Gas' if 'Natural Gas' in x else 'Other')
```

```python
natural_gas = df[df['Natural Gas'] == 'Natural Gas']
```

```python
natural_gas.head()
```

```python
natural_gas[natural_gas['withdrawal_volume_million_gallons'].notna()]
```

```python
import numpy as np
```

```python
natural_gas['withdrawal_volume_million_gallons'] = natural_gas['withdrawal_volume_million_gallons'].replace('--', np.nan)
```

```python
natural_gas[natural_gas['withdrawal_volume_million_gallons'].notna()]
```

```python
natural_gas['discharge_volume_million_gallons'] = natural_gas['discharge_volume_million_gallons'].replace('--', np.nan)
```

```python
natural_gas['consumption_volume_million_gallons'] = natural_gas['consumption_volume_million_gallons'].replace('--', np.nan)
```

```python
water_sources = pd.read_csv('water_sources.csv')
```

```python
water_sources.head()
```

```python
water_sources['plant_name'] = water_sources['Plant Name']
```

```python
water_sources = water_sources.drop(columns='Plant Name')
```

```python
natural_gas = pd.merge(natural_gas, water_sources, on='plant_name', how='left')
```
