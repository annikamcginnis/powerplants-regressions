## Aggregating data by GEOID, keeping only one row per GEOID and creating new columns with all the plant names in that GEOID and the total number of plants

```python
# Step 1: Aggregate data by GEOID
agg_df = df.groupby('GEOID').agg({
    'contains_energy_source': 'sum',
    'plant_name': lambda names: ', '.join(sorted(set(str(n).strip() for n in names if pd.notna(n)))),
}).rename(columns={
    'contains_energy_source': 'num_energy_sources',
    'plant_name': 'plant_names'
}).reset_index()
```

```python
# Step 2: Deduplicate original df by GEOID based on Category_Number == 2 preference
def pick_best_row(group):
    cat2_rows = group[group['Category_Number'] == 2]
    if not cat2_rows.empty:
        return cat2_rows.iloc[0]
    else:
        return group.iloc[0]

deduped_rows = df.groupby('GEOID').apply(pick_best_row).reset_index(drop=True)
```

```python
# Step 3: Merge aggregate info back into deduplicated rows
final_df = pd.merge(deduped_rows, agg_df, on='GEOID', how='left')
```
